### Membership Operators

```python
1 in [1,2,3,4,5]
```

```python
10 not in [1,2,3,4,5]
```

```python
'cat' in 'my pet cat'
```

be careful when using this trick in Python when working with strings as it is not just simply looking for a stand-alone word, but rather just the same sequence
