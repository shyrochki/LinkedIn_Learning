A Python class is a great way to keep related collection of clasess organised. Here and only in classes you can use an uppercase.

## Classes

```python
class Dog:
    def __init__(self, name):
        self.name = name
        self.legs = 4
#initialisation of the function
    def speak(self):
        print(self.name + ' says: Bark')
#the fact that it uses attribution 'self' means that it can access the previous functions or attributes 
#the function (speak) is called a method
```

```python
my_dog = Dog('Rover')
another_dog = Dog('Fluffy')
#this two are instances to the class Dog
#the instance is later (when dot syntax function is used) passed in as the first variable -> object oriented programming
```

```python
my_dog.speak()
```

```python
another_dog.speak()
```
