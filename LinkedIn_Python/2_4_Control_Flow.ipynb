# Control Flow

## If / Else statements

everything indented belongs to the if statement before that, but if it is not indented, the statement does not belong to the if statemnt 
-> that is how you deferentiate to where each statement belongs.

```python
a = False
if a: 
    print('It is true!')
    print('Also print this!')
else:
    print('this is false')
print('Always print this')
```

```python
a = True
b = True 
c = True
if a: 
    print('It is true!')
    print('Also print this!')
    if b: 
        print('Both are true!')
        if c:
            print('All three are true')
else:
    print('this is false')
print('Always print this')
```

```python
a = False
b = True 
c = True
if a: 
    print('It is true!')
    print('Also print this!')
    if b: 
        print('Both are true!')
        if c:
            print('All three are true')
else:
    print('this is false')
print('Always print this')
```

If the first variable is False, the statements written under the first if statement will not go along the way.

## For loops

```python
a = [1,2,3,4,5]
for item in a:
    print (item)
```

"in" in this case is differnet from "in" in Boolean, where 4 in a would give you "True".
The "item" here is just a variable that you are declaring, you can use anything in here, just remember to change the name of the varibale inside the for loop, so that you do not get an error.

## While loops

```python
a = 0
while a < 5:
    print(a)
    a = a + 1
```

You wanna use while loops when the conditions are changing or when they are being impacted by the block under the while loop.
