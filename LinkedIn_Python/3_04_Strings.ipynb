## The uses of f-strings

```python
#Formatting numbers with commas

salary=1000000
f"Salary: {salary:,}"
```

```python
#Percentages

score = 0.85
f'Your score is {score:.0%}'
```

```python
#Padding and alignment

f'{'right':>10}'
```

```python
f'{'left':<10}'
```

```python
f'{'centre':^10}'
```

```python
#Conditional inside f-string

age = 25
f'You are {'adult' if age >= 18 else 'minor'}'
```
