## Boolean Logic

```python
weatherIsNice = True
haveUmbrella = False

if not(haveUmbrella or weatherIsNice):
    print('Stay inside!')
else: 
    print ('Go for a walk')
```

```python
weatherIsNice = False
haveUmbrella = True 

if weatherIsNice or haveUmbrella:
    print('Go for a walk')
else:
    print('Stay inside')
```
